```python

#32.Which age group received the most doses?
male_doses = df['Male (Doses Administered)'].sum()
female_doses = df['Female (Doses Administered)'].sum()
transgender_doses = df['Transgender (Doses Administered)'].sum()

doses_summary = {
    'Male': male_doses,
    'Female': female_doses,
    'Transgender': transgender_doses
}

highest_gender = max(doses_summary, key=doses_summary.get)


print(doses_summary)
print(f"\nThe gender with the highest number of doses administered is: {highest_gender} with {doses_summary[highest_gender]} doses.")
```

```python

#33.getting ratio of male to female and fmale to male
total_male_vaccinated = df['Male(Individuals Vaccinated)'].sum()
total_female_vaccinated = df['Female(Individuals Vaccinated)'].sum()


ratio_male_to_female = total_male_vaccinated / total_female_vaccinated
ratio_female_to_male = total_female_vaccinated/total_male_vaccinated

print(f"Total Male Individuals Vaccinated: {total_male_vaccinated}")
print(f"Total Female Individuals Vaccinated: {total_female_vaccinated}")
print(f"Ratio of Male to Female Individuals Vaccinated: {ratio_male_to_female}")
print(f"ratio of female to female individuals vaccinated:{ratio_female_to_male}")
```
